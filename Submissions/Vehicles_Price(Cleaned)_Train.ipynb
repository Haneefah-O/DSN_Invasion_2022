Converting Objects to numeric data types

```python
#Year and Distance are objects 

veh_price['Year'] = veh_price['Year'].str.replace(',', '') #Removes the comma in the years
veh_price['Year'] = pd.to_numeric(veh_price['Year'])    #converts the column to numeric values

veh_price['Distance'] = veh_price['Distance'].str.replace(',', '') #Removes the comma in the years
veh_price['Distance'] = pd.to_numeric(veh_price['Distance'])    #converts the column to numeric values

veh_price.info()
```

Filling Missing Values

```python
#To fill missing Values in Year and Distance Column

mode_yr = veh_price['Year'].mode()
mid_yr = veh_price['Year'].median() #Getting the median value for Year Column
print(mid_yr)

veh_price['Year'].fillna(mid_yr, inplace=True) #Filling NaN with Median value 
veh_price['Year'].isnull().sum()  #Checking for any missing values in year column

mean_dist = veh_price['Distance'].mean() #Getting mean value for distance
print(mean_dist)

veh_price['Distance'].fillna(mean_dist, inplace=True) #Filling NaN with mean  Value
veh_price['Distance'].isnull().sum() #Checking for missing values in distance column

veh_price.info()
```

```python
#Filling Missing values in Amount and Type Columns

mode_type = veh_price['Type'].mode()
print(mode_type)
mod_typ = str(mode_type)

veh_price['Type'] = veh_price['Type'].fillna(mod_typ)
veh_price['Type'].isnull().sum()

mean_amt = veh_price['Amount (Million Naira)'].mean() #Getting mean value for Amount
print(mean_amt)
veh_price['Amount (Million Naira)'].fillna(mean_amt, inplace=True)
veh_price['Amount (Million Naira)'].isnull().sum()

veh_price.info()
```

```python
veh_price.describe()
```

Dropping Columns

```python
#Check Maker and Model columns to see if we should drop them

veh_price['Maker'].value_counts()
```

```python
veh_price.drop('Model', axis=1, inplace=True)
```

```python
#To turn the objects to category so Machine Learning Model
cat_features = ['Location', 'Maker', 'Colour', 'Type']

for cat_feature in cat_features:
    veh_price[f'{cat_feature}_cat'] = veh_price[cat_feature].astype('category')
    veh_price[f'{cat_feature}_cat'] = veh_price[f'{cat_feature}_cat'].cat.codes

veh_price.head()
```

```python
veh_price.drop(cat_features, axis = 1, inplace=True)
veh_price.head()
```

```python
veh_price2 = veh_price.set_index('VehicleID')
veh_price2.head()
```

```python
veh_price.head()
```

Statistical Models

```python
veh_price2.to_csv('Trained_Cleaned.csv')
```

```python
#Select Target for models
y = veh_price2["Amount (Million Naira)"] # Target
X = veh_price2.drop("Amount (Million Naira)", axis=1)

cleaned_test = pd.read_csv('Test_cleaned.csv')
test_data = cleaned_test.drop('VehicleID', axis=1)
```

```python
#Using Linear Regression
from sklearn.linear_model import LinearRegression
Lin_reg =LinearRegression(positive=True)
Lin_reg.fit(X, y)

y_pred = Lin_reg.predict(test_data)
```

```python
Lin_Reg_Sub = pd.DataFrame({'VehicleID':cleaned_test['VehicleID'], 'Amount (Million Naira)':y_pred})
Lin_Reg_Sub.set_index('VehicleID', inplace=True)
Lin_Reg_Sub.head()
```

```python
Lin_Reg_Sub.to_csv('First Submission(LiR).csv')
```

Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
dt_reg = DecisionTreeRegressor()
dt_reg.fit(X, y)
dt_pred = dt_reg.predict(test_data)
```

```python
dt_reg_sub = pd.DataFrame({'VehicleID':cleaned_test['VehicleID'], 'Amount (Million Naira)':dt_pred})
dt_reg_sub.set_index('VehicleID', inplace=True)
dt_reg_sub.head()
```

```python
dt_reg_sub.to_csv('Second Submission(DT).csv')
```

Random Forest Regression

```python
from sklearn.ensemble import RandomForestRegressor

R_reg = RandomForestRegressor()
R_reg.fit(X, y)
R_pred = R_reg.predict(test_data)

R_reg_sub = pd.DataFrame({'VehicleID':cleaned_test['VehicleID'], 'Amount (Million Naira)':R_pred})
R_reg_sub.set_index('VehicleID', inplace=True)
R_reg_sub.head()
```

```python
R_reg_sub.to_csv('Third Submission(Random Forest).csv')
```
